```python
import pandas as pd
```

```python
df = pd.read_excel('../vacdata/vacancies_2023-04-19.xlsx')
```

```python
gr = df[['job', 'salary_from', 'salary_to', 'salary_apr']].groupby('job')
```

```python
gr.describe().round(0).astype('Int64').to_excel('../vacdata/stat.xlsx')
```

```python
gr.groups.keys()
```

```python
gr.get_group('HR менеджер').describe().round(0).astype('Int64')
```

```python
for group in gr.groups.keys():
    print(group)
    print(gr.get_group(group).describe().round(0), '\n')
```
